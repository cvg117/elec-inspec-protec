```python
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 12]  

# House shapefiles
wi_house_shp = 'data/geo/wi_2018/wi_2018.shp'
pa_house_shp = 'data/geo/pa_2018/pa_2018.shp'
mn_house_shp = 'data/geo/mn_2018/mn_2018.shp'
mi_house_shp = 'data/geo/mi_2018/mi_2018.shp'

# Census 2020 VTD shapefiles
wi_vtd_shp = 'data/geo/tl_2020_55_vtd20/tl_2020_55_vtd20.shp'
pa_vtd_shp = 'data/geo/tl_2020_42_vtd20/tl_2020_42_vtd20.shp'
mn_vtd_shp = 'data/geo/tl_2020_27_vtd20/tl_2020_27_vtd20.shp'
mi_vtd_shp = 'data/geo/tl_2020_26_vtd20/tl_2020_26_vtd20.shp'
```

```python
mn_house_gdf = gp.read_file(mn_house_shp)
mn_house_gdf[['COUNTYNAME', 'COUNTYFIPS', 'VTDID', 'PCTNAME', 'PCTCODE', 'SHORTLABEL',
       'MCDNAME', 'MCDCODE', 'MCDFIPS', 'MCDGNIS']]
```

```python
# Use 'PCTNAME' to match with column 'precinct' in house data
mn_house_gdf['PCTNAME'] = mn_house_gdf['PCTNAME'].str.upper()
```

```python
wi_house_gdf = gp.read_file(wi_house_shp)
wi_house_gdf[['OBJECTID_1', 'CNTY_FIPS', 'CNTY_NAME', 'COUSUBFP', 'MCD_FIPS',
       'MCD_NAME', 'WARDID', 'LABEL']]
```

```python
mi_vtd_gdf = gp.read_file(mi_vtd_shp)
mi_house_gdf = gp.read_file(mi_house_shp)
```

```python
mi_house_gdf.sort_values(by = ['PRECINCTID'], inplace = True)
mi_house_gdf.head()
```

```python
# Only interested in a subset of columns for the purpose of matching
to_keep = ['OBJECTID', 'PRECINCTID', 'COUNTYFIPS', 'MCDFIPS', 'WARD',
       'PRECINCT', 'ShapeSTAre', 'ShapeSTLen', 'precinct_2', 'geometry']
mi_house_gdf = mi_house_gdf[to_keep]
```

```python
#Exploring whether COUNTYFIPS + MCDFIPS + WARD + PRECINCT in house matches NAME20 in census
mi_house_gdf['NAME20'] = mi_house_gdf['COUNTYFIPS'] + mi_house_gdf['MCDFIPS'] + mi_house_gdf['WARD'] + mi_house_gdf['PRECINCT']
mi_house_gdf.sort_values(by=['NAME20'], inplace = True)
mi_house_gdf[['NAME20', 'COUNTYFIPS', 'MCDFIPS', 'WARD', 'PRECINCT']]
```

```python
mi_house_gdf['NAME20_dup'] = mi_house_gdf.duplicated(subset=['NAME20'], keep = False)
mi_house_gdf[mi_house_gdf['NAME20_dup'] == True]
```

```python
mi_vtd_gdf.head()
```

```python
# Drop rows where VTD has ZZZ
print(mi_vtd_gdf.shape)
remove = mi_vtd_gdf['GEOID20'].str.contains('Z')
mi_vtd_gdf = mi_vtd_gdf[ ~remove]
print(mi_vtd_gdf.shape)
```

```python
mi_vtd_gdf.sort_values(by = ['NAME20'], inplace = True)
mi_vtd_gdf['NAME20_dup'] = mi_vtd_gdf.duplicated(subset=['NAME20'], keep = False)
mi_vtd_gdf[mi_vtd_gdf['NAME20_dup'] == True]
```

```python
# Note: 
# Census geometry and house geometry have different CRS
# need to reproject house
mi_vtd_gdf.crs
```

```python
mi_house_gdf.crs
```

```python
# Reproject house boundary to match census
epsg_crs = 'EPSG:4269'
mi_house_rp_gdf  = mi_house_gdf.to_crs(epsg_crs)
```

```python
mi_vtd_precinct = mi_vtd_gdf.merge(mi_house_rp_gdf, on = 'NAME20')
mi_vtd_precinct.head()
mi_to_export
```

```python
 mi_vtd_precinct.shape
```

```python
mi_vtd_precinct.columns
```

```python
vtd = gp.GeoSeries(mi_vtd_precinct['geometry_x'])
house = gp.GeoSeries(mi_vtd_precinct['geometry_y'])
```

```python
vtd.overlaps(house, align = True).value_counts()
```

```python
vtd.intersects(house, align = True).value_counts()
```

```python
mi_vtd_precinct_geo = gp.sjoin(mi_vtd_gdf, mi_house_rp_gdf, how = 'left', predicate = 'overlaps')
```

```python
mi_vtd_precinct_geo.head()
```

```python
mi_vtd_precinct_geo.shape, mi_vtd_precinct.shape
```

```python
# Compare results of merge on 'NAME20' with results of sjon overlaps
mi_comparison = mi_vtd_precinct.merge(mi_vtd_precinct_geo, on = 'GEOID20')
```

```python
mi_comparison.columns
```

```python
mi_comparison['match'] = mi_comparison['PRECINCTID_x'].eq(mi_comparison['PRECINCTID_y'])
mi_comparison['match'].value_counts()
```

```python
mi_comparison[['GEOID20', 'PRECINCTID_x', 'PRECINCTID_y', 'precinct_2_x', 'precinct_2_y', 'match']]
```

```python
mi_to_export = mi_comparison[mi_comparison['match'] == True]
```

```python
mi_to_export.columns
```

```python
# To match with House csv we could use jurisdiction_name or jurisdiction_fips
# jurisdiction_name
# jurisdiction_fips
mi_to_export['jurisdiction_name'] = mi_to_export['precinct_2_x'].str.split('-', 3, expand = True)[1]
mi_to_export['jurisdiction_fips'] = mi_to_export['STATEFP20_x'] + mi_to_export['COUNTYFIPS_x'] + mi_to_export['MCDFIPS_x']
mi_to_export[['jurisdiction_name', 'jurisdiction_fips']]
```

```python
mi_to_export.columns
```

```python
keep = ['STATEFP20_x', 'COUNTYFP20_x', 'VTDST20_x', 'GEOID20', 'VTDI20_x',
       'NAME20', 'NAMELSAD20_x', 'LSAD20_x', 'MTFCC20_x', 'FUNCSTAT20_x', 
       'OBJECTID_x', 'PRECINCTID_x', 'COUNTYFIPS_x',
       'MCDFIPS_x', 'WARD_x', 'PRECINCT_x', 'jurisdiction_name', 'jurisdiction_fips']
```

```python
mi_to_export.to_csv('data/mi_match_vtd_house.csv', columns = keep, index = False)
```
